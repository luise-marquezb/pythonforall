#1. Matriz

##1.1 Creación de Matriz

```python
matriz = []

FILAS = 4
COLUMNAS = 6

#Llenar matriz
for i in range(FILAS) : # 4 veces
    fila = [0] * COLUMNAS # [0,0,0,0,0,0]    
    matriz.append(fila)



#Mostrar matriz

print(matriz)  

#Recorrer matriz
print("Recorrido de matriz por medio de FOR")

for i in range(FILAS): # repite 4 veces
    for j in range(COLUMNAS): # repite 6 veces 
        print(matriz[i][j], end="  ")
    print()



```

##1.4 Sumatoria de filas

```python
TRIMESTRES = 3  # columnas
DIVISIONES = 5  # filas

suma = 0


divisiones = ["Línea Blanca", "Electrodomésticos", "Juguetería",
              "Perecibles", "Limpieza"]
datos = [[450, 650, 342], [340, 487, 767], [134, 212, 354],
         [180, 464, 565], [647, 324, 232]]  # matriz es una lista de listas


i = 2
for j in range(TRIMESTRES):  # j =  0, 1, 2. Se repite 3 veces 
    suma = suma + datos[i][j]  
   
    # datos[0][0] = 450 
    # suma = 0 + 450 = 450

    # datos[0][1] = 650
    # suma = 450 + 650 = 1100

    # datos[0][2] = 342
    # suma = 1100 + 342 = 1442

print("\nEl total de ventas de la División Juguetería es: ",suma)
```

##1.5 Sumatoria de Columnas.

```python
TRIMESTRES = 3  # columnas
DIVISIONES = 5  # filas

suma = 0
divisiones = ["Línea Blanca", "Electrodomésticos", "Juguetería",
              "Perecibles", "Limpieza"]
datos = [[450, 650, 342], [340, 487, 767], [134, 212, 354],
         [180, 464, 565], [647, 324, 232]] # matriz es una lista de listas



j = 0
for i in range(DIVISIONES) : # i  = 0 , 1, 2, 3, 4. Se repite 5 veces
    suma = suma + datos[i][j]  
    # datos[0][0]  = 450
    # datos[1][0]  = 340
    # datos[2][0]  = 134
    # datos[3][0]  = 180
    # datos[4][0]  = 647
print("\nEl total de ventas del primer Trimestre es: ", suma)
```

#2. Matrices & Graphics Computing

```python
import numpy as np
import matplotlib.pyplot as plt

fils = 255
cols = 255
matrix = np.zeros([fils, cols, 3], dtype=np.uint8)


for i in range(fils):
   for j in range(cols):
       matrix[i][j] = [0, 0, 50] # [rojo, verde, azul]

plt.imshow(matrix)
```

```python
import numpy as np
import matplotlib.pyplot as plt
matrix = np.zeros([100, 300, 3], dtype=np.uint8)


matrix[:,:100] = [0, 10, 0]                # Rojo
matrix[:,100:200] = [255, 255, 255]   # Blanco
matrix[:,200:] = [255, 0, 0]                # Rojo
plt.imshow(matrix)
```
